```python
# 使用pyecharts画出柱状图
from pyecharts.charts import Bar

bar = Bar()
# 添加x轴和y轴数据
bar.add_xaxis(["衬衫","羊毛衫","雪纺衫","衬衫","羊毛衫","雪纺衫"])
bar.add_yaxis("商家A",[5,20,36,10,75,90])
# render 会生成本地 HTML 文件，默认会在当前目录生成 render.html 文件
# 也可以传入路径参数，如 bar.render("mycharts.html")
# bar.render("mycharts.html") # '/home/toohoo/PycharmProjects/DataAnalysis/DataAnalysisInAction/lesson15/mycharts.html'
bar.render_notebook()
```

```python
# 默认的主题是红黑的
from pyecharts.charts import Bar
# 或者更换一下主题
from pyecharts import options as opts
from pyecharts.globals import ThemeType

bar = Bar(init_opts = opts.InitOpts(theme = ThemeType.WESTEROS)) # 更改主题为深蓝和浅蓝
# 添加x轴和y轴数据
bar.add_xaxis(["衬衫","羊毛衫","雪纺衫","衬衫","羊毛衫","雪纺衫"])
bar.add_yaxis("商家A",[5,20,36,10,75,90])
bar.add_yaxis("商家B",[15,6,45,20,35,66])
# bar.render("mycharts.html") # '/home/toohoo/PycharmProjects/DataAnalysis/DataAnalysisInAction/lesson15/mycharts.html'
bar.render_notebook()
```

```python
import datetime
import random

from pyecharts import options as opts
from pyecharts.charts import Calendar


def calendar_base() -> Calendar:
    begin = datetime.date(2017, 1, 1)
    end = datetime.date(2017, 12, 31)
    data = [
        [str(begin + datetime.timedelta(days=i)), random.randint(1000, 25000)]
        for i in range((end - begin).days + 1)
    ]

    c = (
        Calendar()
        .add("", data, calendar_opts=opts.CalendarOpts(range_="2017"))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Calendar-2017年微信步数情况"),
            visualmap_opts=opts.VisualMapOpts(
                max_=20000,
                min_=500,
                orient="horizontal",
                is_piecewise=True,
                pos_top="230px",
                pos_left="100px",
            ),
        )
    )
    return c

calendar_base().render_notebook()
```
